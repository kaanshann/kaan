```python
filtered_data = data[(data['Year'] == 2016) & (data['Gender'] == 'F')]
education_counts = filtered_data.groupby('Education Level of Mother')['Number of Births'].sum()

plt.bar(education_counts.index, education_counts.values)
plt.xlabel('Education Level of Mother')
plt.ylabel('Number of Births')
plt.title('Total Number of Births by Education Level (Year: 2016, Gender: Female)')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_age_by_year = data.groupby('Year')['Average Age of Mother (years)'].mean()

plt.plot(avg_age_by_year.index, avg_age_by_year.values)
plt.xlabel('Year')
plt.ylabel('Average Age of Mother (years)')
plt.title('Average Age of Mother Over the Years')
plt.show()
```
